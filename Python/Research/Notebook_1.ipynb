```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
main_df = pd.read_csv("Data_1-sensor-fault-detection.csv", sep=";", parse_dates=['Timestamp'])
main_df
```

```python
dummy_df = main_df
dummy_df.info()
```

```python
dummy_in_df = dummy_df.set_index("Timestamp")
dummy_in_df
```

```python
sigma_over_4 = dummy_in_df["Value"].mean() + (4 * dummy_df["Value"].std())
sigma_under_4 = dummy_in_df["Value"].mean() - (4 * dummy_df["Value"].std())
dummy_in_df['Outside'] = np.where((dummy_df.Value < sigma_over_4) & (dummy_df.Value > sigma_under_4), False, True)
dummy_in_df
```
