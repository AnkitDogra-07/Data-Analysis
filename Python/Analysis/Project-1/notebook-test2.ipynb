```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("data-IPL_Matches_2008_2022.csv")
df
```

```python
df.columns
```

```python
modified_df = df.drop(columns=['SuperOver', 'method', 'Team1Players','Team2Players'])
modified_df.columns
```

```python
modified_df
```

```python
grp_df_season = df.groupby(['Season'])['Team1'].nunique()
grp_df_season = pd.DataFrame(grp_df_season)
grp_df_season
```

```python
grp_df_season = grp_df_season.rename(columns = {'Team1':'Total_Teams'}).reset_index()
grp_df_season
```

```python
grp_df_matches = df.groupby(['Season'])['MatchNumber'].agg('count')
grp_df_matches = pd.DataFrame(grp_df_matches)
grp_df_matches
```

```python
grp_df_matches = grp_df_matches.rename(columns = {'MatchNumber':'Total_Matches'}).reset_index()
grp_df_matches
```

```python
grp_df_matches
```

```python
new_df = pd.merge(left=grp_df_matches, right=grp_df_season, how='left', left_on='Season', right_on='Season')
new_df
```
