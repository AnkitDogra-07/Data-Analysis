```python
import pandas as pd
```

```python

df = pd.read_csv("data_cleaning_challenge.csv")
df
```

```python
(df[df.columns].isnull().sum()/df.shape[0])*100
```

```python
df.columns[9:]
```

```python
mod_df = df.drop(columns=list(df.columns[9:]))
mod_df
```

```python
mod_df_nan = mod_df[mod_df["Row Type"].notna()]
mod_df_nan
```

```python
column_values = []
counter = 0

for i in mod_df_nan["Row Type"]:
    if "first name" in i:
        counter+=1
    column_values.append(counter)
column_values
```

```python
mod_df_iter = mod_df_nan
mod_df_iter['Iteration'] = column_values
mod_df_iter
```

```python
mod_df_extra = mod_df_iter[mod_df_iter["Row Type"] != "Row Type"]
mod_df_extra
```

```python
mod_df_extra = mod_df_extra.reset_index()
```

```python
mod_df_extra.drop(columns=['index'], inplace=True)
mod_df_extra
```

```python
mod_df_new = mod_df_extra[mod_df_extra["Row Type"].str.contains("first name")]
mod_df_new
```

```python
mod_df_new_extra = mod_df_new.drop(columns=mod_df_new.columns[3:-1])
mod_df_new_extra
```

```python
mod_df_new_extra.rename(columns={"Row Type":"First Name", "Iter Number":"Last Name", "Power1":"Date"}, inplace=True)
mod_df_new_extra
```

```python
mod_df_new_extra["First Name"] = mod_df_new_extra["First Name"].str[12:]
mod_df_new_extra["Last Name"] = mod_df_new_extra["Last Name"].str[11:]
mod_df_new_extra["Date"] = mod_df_new_extra["Date"].str[6:]
mod_df_new_extra = mod_df_new_extra.reset_index()
mod_df_new_extra
```

```python
mod_df_new_extra.drop(columns=['index'], inplace=True)
mod_df_new_extra
```

```python
new_mod_df = mod_df_extra[~(mod_df_extra["Row Type"].str.contains("first name"))]
new_mod_df
```

```python
new_mod_df = new_mod_df.reset_index()
new_mod_df.drop(columns=['index'], inplace=True)
new_mod_df
```

```python
final_df = pd.merge(left=mod_df_new_extra, right=new_mod_df, left_on="Iteration", right_on="Iteration", how="left")
final_df
```
